# Variable stattistics

## Visualize the distribution of variables

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(filepath_or_buffer='../data/interim/handled_missing_values.csv')
```

```python
data.hist(bins= 25, figsize=(15, 10))
```

## Visualize categorical variables


```python
plt.figure(figsize=(15, 15))

# Columns with object or category data type
categorical_cols = data.select_dtypes(include=['object', 'category']).columns

# Calculate the number of rows and columns for subplots
total_cols = len(categorical_cols)
cols = 4  # Number of columns in subplot grid
rows = total_cols // cols + (total_cols % cols > 0)

# Plot bar plots for categorical columns
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(rows, cols, i)
    sns.countplot(x=col, data=data, hue=col,palette='pastel')  # i bng mu thnh 'pastel'
    plt.title(col, fontweight='bold')
    plt.xticks(rotation=25)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```
